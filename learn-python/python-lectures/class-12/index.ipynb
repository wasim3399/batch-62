# Error Handling

```python
a : list = [1, 2, 3, 4, 5]

# Dynamic error handling
print('logic 1')
print('logic 2')
try:
    print(5/2)
    print(a[2])
    open(myFile.txt)
except ZeroDivisionError:
    print('zero division error')
except IndexError:
    print('index error')
except: # unhandled exception
    print('unknown error')
print('logic 3')
print('logic 4')
print('logic 5')
```

```python
a : list = [1, 2, 3, 4, 5]

# Dynamic error handling
print('logic 1')
print('logic 2')
try:
    print(5/2)
    print(a[2])
    open(myFile.txt)
except ZeroDivisionError:
    print('zero division error')
except IndexError:
    print('index error')
except Exception as e: # unhandled exception with error message
    print(f'unknown error {e}')
print('logic 3')
print('logic 4')
print('logic 5')
```

# Custom error messaging

```python
class StudentCard():
    def __init__(self, name : str, age : int) -> None:
        if (age <= 0):
            raise StudentAgeError('Student age error')
        
        self.name = name
        self.age = age

class StudentAgeError(Exception):
    pass

student1 = StudentCard('wasim', -3)
```

```python
from typing import TextIO
data : TextIO = open('file.txt') # connect with the file
print(data.read())
data.close() # close the connection
```

```python
# 'with' auto close the connection when you come out the block
with open('file.txt') as file:
    print(file.read())

print('hello world')
```

```python
with open('file.txt') as file:
    print(file.readline())

print('hello world')
```

```python
with open('file.txt') as file:
    print(file.readlines())

print('hello world')
```

```python
with open('file.txt') as file:
    print(file.readlines()[:2])

print('hello world')
```

```python
# check file permission table
# check file modes

with open('test.txt', 'w') as file:
    print(file.read())
```

```python
# cursor position is very important when perform operations on the file and goes to the end of the file

with open('test.txt', 'r+') as file:
    print(file.read())
    print(file.write('\n Hello World'))
    file.seek(0) # moves the cursor to the beginning of the file

    print('after read', file.read())
```
