## Add custom function on series data

```python
def grade(per : float) -> str: # type: ignore
    if(per > 90):
        return 'A+'
    elif(per >= 80 and per < 90):
        return 'A'
    elif(per >= 70  and per < 80):
        return 'B'
    else:
        return 'F'
    
grade(43)
```

```python
df['grade'] = df['percentage'].apply(grade)
df
```

```python
import pandas as pd
import json

# Load the CSV file
file_path = 'query_result.csv'  # Update with your file's path
# Use 'on_bad_lines="skip"' to ignore problematic rows
df = pd.read_csv(file_path, on_bad_lines="skip")

# Assuming all JSON-like data is in a column named "request_data"
json_data = []

for index, row in df.iterrows():
    try:
        # Attempt to parse the JSON content in each row
        json_row = json.loads(row['request_data'])
        json_data.append(json_row)
    except (json.JSONDecodeError, KeyError):
        # Skip rows that can’t be parsed
        continue

# Save to JSON file
output_file_path = 'output_data.json'
with open(output_file_path, 'w') as json_file:
    json.dump(json_data, json_file, indent=4)

print(f"Data successfully converted and saved to {output_file_path}")
```
